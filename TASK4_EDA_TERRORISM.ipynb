```python
df['Wounded'] = df['Wounded'].fillna(0).astype(int)
df['Killed'] = df['Killed'].fillna(0).astype(int)
df['casualities'] = df['Killed'] + df['Wounded']
```

```python
df1=df.sort_values(by='casualities',ascending=False)[:40]
```

```python
heat=df1.pivot_table(index='Country',columns='Year',values='casualities')
heat.fillna(0, inplace=True)
heat.head()
```

```python
import plotly.offline as py
import plotly.graph_objs as go
heatmap = go.Heatmap(z=heat.values, x=heat.columns, y=heat.index)
data = [heatmap]
layout = go.Layout(
    title='Top 40 Worst Terror Attacks in History from 1982 to 2016',
    xaxis = dict(ticks='', nticks=20),
    yaxis = dict(ticks='')
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='heatmap',show_link=False)
```

```python
killd=df.loc[:,'Killed']
print('Number of people killed by terror attack:',int(sum(killd.dropna())))
```

```python
attackd=df.loc[:,'AttackType']
typekilldata=pd.concat([attackd,killd],axis=1)
```

```python
typekilldata.tail()
```

```python
typeKillFormatData = typekilldata.pivot_table(columns='AttackType', values='Killed', aggfunc='sum')
typeKillFormatData
```

```python
typeKillFormatData.info()
```

```python
labels=typeKillFormatData.columns.tolist()
```

```python
summed_values = typeKillFormatData.sum(axis=0)
values = summed_values.values
fig, ax = plt.subplots(figsize=(20, 20), subplot_kw=dict(aspect="equal"))
ax.pie(values, startangle=90, autopct='%.2f%%')
plt.title('Types of terrorist attacks that cause deaths')
plt.legend(labels, loc='upper right', bbox_to_anchor=(1.3, 0.9), fontsize=15)
plt.show()
```

```python
#armed assault and bombing/explosion are the major causes of death in
#these attacks.
```

```python
countryd=df.loc[:,'Country']
countrykilld=pd.concat([countryd,killd],axis=1)
countrykilld
```

```python
countryKillFormatData = countrykilld.pivot_table(columns='Country', values='Killed', aggfunc='sum')
countryKillFormatData
```

```python
transpoze = countryKillFormatData.T
values = transpoze.sum(axis=1).reset_index().values.tolist()
values.sort(key=lambda x: x[1], reverse=True)
top_20_values = values[:20]
labels = [x[0] for x in top_20_values]
values = [int(x[1]) for x in top_20_values]
index = np.arange(len(labels))
plt.bar(index, values)
plt.ylabel('Killed People', fontsize=20)
plt.xlabel('Countries', fontsize=20)
plt.xticks(index, labels, fontsize=18, rotation=90)
plt.title('Number of people killed by countries', fontsize=20)
plt.show()
```
